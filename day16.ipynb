```python
import seaborn as sns
import matplotlib.pyplot as plt
iris_data=sns.load_dataset('iris')
iris_data
```

```python
sns.boxplot(x=iris_data['sepal_length'],color='purple')
```

```python
sns.boxplot(x=iris_data['sepal_width'],palette=['violet','red','green'],y=iris_data['species'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='species', y='sepal_width', data=iris_data, palette='Set2', hue='species')
plt.show()

```

```python
df=sns.load_dataset('taxis')
sns.scatterplot(x=df['fare'],y=df['total'],palette=['black','red'],hue=df['payment'])
plt.show()
```
